```python
import pandas as pd
import numpy as np

# Dosya yollarını kendi bilgisayarındaki klasör yoluna göre güncelle
feature_file = r"test\Partition1_LSBZM-Norm_FPCKNN-impute.pkl"
label_file = r"test\Partition1_Labels_LSBZM-Norm_FPCKNN-impute.pkl"

print("NumPy matrisleri yükleniyor, lütfen bekleyin...")

# pd.read_pickle hala çalışır çünkü pickle içindeki her türlü Python nesnesini çıkarabilir
features_array = pd.read_pickle(feature_file)
labels_array = pd.read_pickle(label_file)

print("\n--- ÖZELLİKLER (FEATURES) ---")
print(f"Tip: {type(features_array)}")
# En çok merak ettiğimiz yer burası! (Örnek Sayısı, Zaman Adımı, Özellik Sayısı) mı göreceğiz?
print(f"Boyut (Shape): {features_array.shape}") 

print("\n--- ETİKETLER (LABELS) ---")
print(f"Tip: {type(labels_array)}")
print(f"Boyut (Shape): {labels_array.shape}")
print("İlk 15 etiket:\n", labels_array[:15])

# İçerideki sınıfları bulalım (X, M, C, B mi yazıyor yoksa 0 ve 1 mi?)
benzersiz_etiketler = np.unique(labels_array)
print(f"\nVeri setindeki tüm farklı etiket sınıfları: {benzersiz_etiketler}")
```

```python
import numpy as np
import pandas as pd
import torch
from torch.utils.data import TensorDataset, DataLoader

# 1. Dosyaları Yükleme Fonksiyonu
def load_partition(partition_num):
    # Dosya yollarının senin bilgisayarındakiyle aynı olduğundan emin ol
    feature_file = f"test\\Partition{partition_num}_LSBZM-Norm_FPCKNN-impute.pkl"
    label_file = f"test\\Partition{partition_num}_Labels_LSBZM-Norm_FPCKNN-impute.pkl"
    
    print(f"Partition {partition_num} yükleniyor...")
    X = pd.read_pickle(feature_file)
    y = pd.read_pickle(label_file)
    return X, y

# 2. Eğitim (Train) Verisi: Partition 1, 2, 3 birleştirilir
X_p1, y_p1 = load_partition(1)
X_p2, y_p2 = load_partition(2)
X_p3, y_p3 = load_partition(3)

# Matrisleri alt alta ekliyoruz (Satır sayısı artacak)
X_train = np.concatenate((X_p1, X_p2, X_p3), axis=0)
y_train = np.concatenate((y_p1, y_p2, y_p3), axis=0)

# 3. Validation (Doğrulama) Verisi: Partition 4
X_val, y_val = load_partition(4)

# 4. Test Verisi: Partition 5
X_test, y_test = load_partition(5)

print("\n--- VERİ SETİ BAŞARIYLA BİRLEŞTİRİLDİ ---")
print(f"Eğitim Seti (Train): X={X_train.shape}, y={y_train.shape}")
print(f"Doğrulama Seti (Val): X={X_val.shape}, y={y_val.shape}")
print(f"Test Seti (Test): X={X_test.shape}, y={y_test.shape}")

# 5. PyTorch Tensörlerine Dönüştürme
X_train_t = torch.tensor(X_train, dtype=torch.float32)
y_train_t = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)

X_val_t = torch.tensor(X_val, dtype=torch.float32)
y_val_t = torch.tensor(y_val, dtype=torch.float32).unsqueeze(1)

X_test_t = torch.tensor(X_test, dtype=torch.float32)
y_test_t = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)

# 6. DataLoader Hazırlığı
# Veri seti çok büyük olduğu için Batch Size'ı 64'ten 128'e çıkardık, böylece eğitim daha hızlı olur
BATCH_SIZE = 128 
train_loader = DataLoader(TensorDataset(X_train_t, y_train_t), batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(TensorDataset(X_val_t, y_val_t), batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(TensorDataset(X_test_t, y_test_t), batch_size=BATCH_SIZE, shuffle=False)

print("\nPyTorch DataLoader'lar eğitime hazır!")
```

```python
# attention class_weights LSTM

# PyTorch LSTM + Attention Mimarisi
import torch.nn as nn
import torch.nn.functional as F

class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super(Attention, self).__init__()
        # Dikkat skorlarını hesaplamak için basit bir lineer katman
        self.attention = nn.Linear(hidden_dim, 1, bias=False)

    def forward(self, lstm_out):
        # lstm_out boyutu: (batch_size, time_steps, hidden_dim)
        
        # Her zaman adımı için bir skor hesapla
        attn_scores = self.attention(lstm_out) # Boyut: (batch_size, time_steps, 1)
        
        # Skorları 0-1 arası olasılıklara (ağırlıklara) çevir
        attn_weights = F.softmax(attn_scores, dim=1) 
        
        # LSTM çıktılarını bu dikkat ağırlıklarıyla çarp ve topla (Context Vector)
        # Hangi zaman adımının ağırlığı yüksekse, model o anki veriye daha çok odaklanır
        context_vector = torch.sum(attn_weights * lstm_out, dim=1) # Boyut: (batch_size, hidden_dim)
        
        return context_vector, attn_weights

class SolarFlarePredictor(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, dropout_rate=0.3):
        super(SolarFlarePredictor, self).__init__()
        
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        
        # LSTM Katmanı
        # batch_first=True çok önemli! Çünkü tensörlerimiz (Batch, Time, Features) sırasında
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, 
                            batch_first=True, 
                            dropout=dropout_rate if num_layers > 1 else 0)
        
        # Attention Katmanımız
        self.attention = Attention(hidden_dim)
        
        # Tam Bağlı (Dense/Linear) Katmanlar
        self.fc1 = nn.Linear(hidden_dim, 32)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout_rate)
        
        # Çıktı katmanı: Sadece 1 nöron (Patlama olacak mı, olmayacak mı?)
        self.fc2 = nn.Linear(32, 1) 
        self.sigmoid = nn.Sigmoid() # Çıktıyı 0 ile 1 arasına (olasılığa) sıkıştırır
        
    def forward(self, x):
        # x'in boyutu: (batch_size, time_steps, input_dim)
        
        # Veriyi LSTM'den geçiriyoruz
        lstm_out, (h_n, c_n) = self.lstm(x)
        
        # LSTM çıktılarını Attention mekanizmasına veriyoruz
        context_vector, attn_weights = self.attention(lstm_out)
        
        # Dikkatlice seçilmiş bilgiyi sinir ağına sokuyoruz
        out = self.fc1(context_vector)
        out = self.relu(out)
        out = self.dropout(out)
        out = self.fc2(out)
        
        # Sonuç: 0 ile 1 arası olasılık
        olasilik = self.sigmoid(out)
        
        return olasilik, attn_weights

# --- MODELİ BAŞLATMA VE TEST ETME ---

# Hiperparametreler
INPUT_DIM = 24       # Veri setimizdeki özellik sayısı (örneğin manyetik akı, alan vb.)
HIDDEN_DIM = 64      # LSTM'in içindeki hafıza hücresi sayısı
NUM_LAYERS = 2       # Üst üste 2 LSTM katmanı

# Modeli yarat
model = SolarFlarePredictor(input_dim=INPUT_DIM, hidden_dim=HIDDEN_DIM, num_layers=NUM_LAYERS)
print(model)
```

```python
import torch.optim as optim
import torch.nn as nn

# 1. Cihaz Kontrolü: GPU (CUDA) var mı, yoksa CPU ile mi devam edeceğiz?
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Eğitim için kullanılacak donanım: {device}")

# 2. Modeli Baştan Yarat ve Cihaza Gönder
# Daha önceki hücrede yazdığımız SolarFlarePredictor sınıfını kullanıyoruz
INPUT_DIM = 24
HIDDEN_DIM = 64
NUM_LAYERS = 2

# Modeli oluşturup direkt cihaza (.to(device)) yolluyoruz
model = SolarFlarePredictor(input_dim=INPUT_DIM, hidden_dim=HIDDEN_DIM, num_layers=NUM_LAYERS).to(device)

# 3. Gerçek Verideki Dengesizlik Oranını Hesapla
# numpy dizisi olduğu için .sum() doğrudan çalışır
neg_sayisi = (y_train == 0).sum()
poz_sayisi = (y_train == 1).sum()
gercek_pos_weight = torch.tensor([neg_sayisi / poz_sayisi], dtype=torch.float32).to(device)

print(f"Gerçek Veride 'Patlama Yok' Sayısı: {neg_sayisi}")
print(f"Gerçek Veride 'Patlama Var' Sayısı: {poz_sayisi}")
print(f"Yeni Ceza Katsayısı (pos_weight): {gercek_pos_weight.item():.2f}")

# 4. Optimizatör ve Kayıp Fonksiyonu
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

EPOCHS = 10 # Veri devasa olduğu için şimdilik 10 tur (epoch) yeterli olacaktır

print("\n--- GERÇEK VERİ İLE EĞİTİM BAŞLIYOR ---")
for epoch in range(EPOCHS):
    model.train()
    toplam_train_loss = 0
    
    for batch_X, batch_y in train_loader:
        # Verileri donanıma (GPU veya CPU) gönder
        batch_X, batch_y = batch_X.to(device), batch_y.to(device)
        
        optimizer.zero_grad()
        
        tahminler, _ = model(batch_X)
        loss = criterion(tahminler, batch_y)
        
        # Gerçek ağırlıklı kaybı hesapla
        agirlikli_loss = loss * torch.where(batch_y == 1, gercek_pos_weight, torch.tensor(1.0).to(device))
        ortalama_loss = agirlikli_loss.mean()
        
        ortalama_loss.backward()
        optimizer.step()
        
        toplam_train_loss += ortalama_loss.item()
        
    # --- Validation (Doğrulama) Aşaması ---
    model.eval()
    toplam_val_loss = 0
    with torch.no_grad():
        for val_X, val_y in val_loader:
            val_X, val_y = val_X.to(device), val_y.to(device)
            val_tahminler, _ = model(val_X)
            
            # Validation loss'u ağırlıksız hesaplıyoruz ki saf performansı görelim
            v_loss = criterion(val_tahminler, val_y)
            toplam_val_loss += v_loss.item()
            
    avg_train_loss = toplam_train_loss / len(train_loader)
    avg_val_loss = toplam_val_loss / len(val_loader)
    
    print(f"Epoch {epoch+1}/{EPOCHS} | Train Loss: {avg_train_loss:.4f} | Validation Loss: {avg_val_loss:.4f}")
```

```python
# Hücre 9 (Güncel): Test, Değerlendirme ve Görselleştirme (GPU Uyumlu)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("Gerçek verilerle test aşaması başlıyor...")

model.eval() # Modeli değerlendirme moduna al
test_tahminleri = []
test_gercekleri = []
ornek_attention_agirliklari = None

with torch.no_grad():
    for batch_X, batch_y in test_loader:
        # 1. Verileri modele girmeden önce cihazda (GPU/CPU) olduğundan emin ol
        batch_X = batch_X.to(device)
        batch_y = batch_y.to(device)
        
        tahmin_olasiliklari, attn_weights = model(batch_X)
        
        tahmin_siniflari = (tahmin_olasiliklari >= 0.5).float()
        
        # 2. HATA ÖNLEYİCİ: Tensörleri numpy dizisine çevirmeden önce .cpu() ile işlemciye çek
        test_tahminleri.extend(tahmin_siniflari.cpu().numpy())
        test_gercekleri.extend(batch_y.cpu().numpy())
        
        # Görselleştirme için attention ağırlıklarını sakla (yine cpu'ya çekerek)
        if ornek_attention_agirliklari is None:
            ornek_attention_agirliklari = attn_weights[0].cpu().numpy()

# 3. Metrikleri Hesapla
acc = accuracy_score(test_gercekleri, test_tahminleri)
prec = precision_score(test_gercekleri, test_tahminleri, zero_division=0)
rec = recall_score(test_gercekleri, test_tahminleri, zero_division=0)
f1 = f1_score(test_gercekleri, test_tahminleri, zero_division=0)

print("\n--- GERÇEK TEST SETİ SONUÇLARI ---")
print(f"Accuracy (Doğruluk):  {acc:.4f}")
print(f"Precision (Kesinlik): {prec:.4f}")
print(f"Recall (Duyarlılık):  {rec:.4f}")
print(f"F1-Score:             {f1:.4f}")

# 4. Görselleştirme (Karmaşıklık Matrisi)
cm = confusion_matrix(test_gercekleri, test_tahminleri)

plt.figure(figsize=(14, 5))

# Sol Grafik: Confusion Matrix
plt.subplot(1, 2, 1)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Patlama Yok', 'Patlama Var'], 
            yticklabels=['Patlama Yok', 'Patlama Var'])
plt.title('SWAN-SF Test Seti Karmaşıklık Matrisi')
plt.xlabel('Modelin Tahmini')
plt.ylabel('Gerçek Durum')

# Sağ Grafik: Attention Ağırlıkları
plt.subplot(1, 2, 2)
# Boyutu (60, 1) olan attention ağırlıklarını düzleştiriyoruz (.flatten())
plt.plot(ornek_attention_agirliklari.flatten(), marker='o', color='darkorange')
plt.title('Örnek Bir Zaman Penceresinde Attention Odakları')
plt.xlabel('Zaman Adımı (Son 12 Saat: Geçmişten -> Şimdiki Zamana)')
plt.ylabel('Dikkat Skoru (Önem Derecesi)')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Hücre 11: Eşik Değeri (Threshold) Optimizasyonu
from sklearn.metrics import precision_recall_curve
import numpy as np
import matplotlib.pyplot as plt

print("Modelin karar eşikleri analiz ediliyor...")

model.eval()
test_olasiliklari = []
test_gercekleri_opt = []

# Modelin ham olasılık (0 ile 1 arası) tahminlerini topluyoruz
with torch.no_grad():
    for batch_X, batch_y in test_loader:
        batch_X = batch_X.to(device)
        tahmin_olasiliklari, _ = model(batch_X)
        test_olasiliklari.extend(tahmin_olasiliklari.cpu().numpy())
        test_gercekleri_opt.extend(batch_y.cpu().numpy())

# Precision-Recall Eğrisi hesaplama
precisions, recalls, thresholds = precision_recall_curve(test_gercekleri_opt, test_olasiliklari)

# F1-Score'u her bir eşik değeri için hesaplıyoruz
# Uyarı: Sıfıra bölme hatasını önlemek için paydanın 0 olduğu yerleri düzeltiyoruz
f1_skorlari = np.divide(2 * (precisions * recalls), (precisions + recalls), 
                        out=np.zeros_like(precisions), where=(precisions + recalls) != 0)

# En yüksek F1 skorunu veren eşik değerini buluyoruz
en_iyi_indeks = np.argmax(f1_skorlari)
en_iyi_esik = thresholds[en_iyi_indeks]
en_iyi_f1 = f1_skorlari[en_iyi_indeks]

print(f"\n--- OPTİMİZASYON SONUÇLARI ---")
print(f"Eski Karar Eşiği: 0.500")
print(f"Mükemmel Denge İçin Yeni Eşik: {en_iyi_esik:.4f}")
print(f"Bu eşikte beklenen F1-Score: {en_iyi_f1:.4f} (Eski F1: 0.3762)")

# Görselleştirme
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precisions[:-1], 'b--', label='Precision (Yanlış Alarmları Azaltma)')
plt.plot(thresholds, recalls[:-1], 'g-', label='Recall (Gerçekleri Yakalama)')
plt.plot(thresholds, f1_skorlari[:-1], 'r-', linewidth=2, label='F1-Score (Genel Başarı)')
plt.axvline(x=en_iyi_esik, color='k', linestyle=':', label=f'En İyi Eşik ({en_iyi_esik:.2f})')
plt.title('Hangi Olasılık Eşiğinde En Mantıklı Kararı Veriyoruz?')
plt.xlabel('Karar Eşiği (Threshold)')
plt.ylabel('Skor')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Yeni Eşik ile Performans Testi
yeni_tahminler = (np.array(test_olasiliklari) >= en_iyi_esik).astype(int)

yeni_acc = accuracy_score(test_gercekleri_opt, yeni_tahminler)
yeni_prec = precision_score(test_gercekleri_opt, yeni_tahminler, zero_division=0)
yeni_rec = recall_score(test_gercekleri_opt, yeni_tahminler, zero_division=0)

print("\n--- YENİ EŞİK DEĞERİ İLE METRİKLER ---")
print(f"Accuracy:  {yeni_acc:.4f}")
print(f"Precision: {yeni_prec:.4f}")
print(f"Recall:    {yeni_rec:.4f}")
```

```python
import torch
import numpy as np

# --- 1. MODELİ YÜKLEME AŞAMASI ---
# Uyarı: Bu kodu çalıştıracağın dosyada SolarFlarePredictor sınıfının önceden tanımlanmış olması gerekir.

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Tahmin (Inference) için kullanılacak donanım: {device}")

# Modeli mimarisiyle sıfırdan oluştur
yuklenen_model = SolarFlarePredictor(input_dim=24, hidden_dim=64, num_layers=2)

# Kaydettiğimiz ağırlıkları içine yükle 
# (map_location=device sayesinde GPU'da eğitilen model CPU'da da sorunsuz açılır)
yuklenen_model.load_state_dict(torch.load('solar_flare_lstm_v1.pth', map_location=device))

# Modeli cihazına gönder ve KESİNLİKLE Değerlendirme (Eval) moduna al
yuklenen_model.to(device)
yuklenen_model.eval() 
print("Model başarıyla yüklendi ve tahmin yapmaya hazır!")

# --- 2. TAHMİN (PREDICT) FONKSİYONU ---

def predict_flare(model, yeni_veri, esik_degeri=0.5):
    """
    Modele yeni veriler vererek patlama ihtimalini hesaplar.
    
    Parametreler:
    - model: Yüklenmiş ve eval moduna alınmış PyTorch modeli
    - yeni_veri: (N, 60, 24) veya tek örnek için (60, 24) boyutlarında NumPy dizisi
    - esik_degeri: Hangi olasılığın üzerini 'Patlama Var' kabul edeceğiz? (Varsayılan 0.5)
    """
    
    # 1. Gelen veri NumPy dizisi ise PyTorch Tensörüne çevir
    if isinstance(yeni_veri, np.ndarray):
        veri_tensor = torch.tensor(yeni_veri, dtype=torch.float32)
    else:
        # Zaten tensörse güvenlik için kopyala ve float yap
        veri_tensor = yeni_veri.clone().detach().float()
        
    # 2. Boyut Kontrolü: Eğer tek bir örnek geldiyse (60, 24)
    # PyTorch'un beklediği Batch boyutunu ekleyerek (1, 60, 24) yap
    if len(veri_tensor.shape) == 2:
        veri_tensor = veri_tensor.unsqueeze(0)
        
    # Veriyi işlem yapılacak cihaza gönder
    veri_tensor = veri_tensor.to(device)
    
    # 3. Tahmin İşlemi (Gereksiz bellek kullanımını önlemek için no_grad)
    with torch.no_grad():
        olasiliklar, attention_agirliklari = model(veri_tensor)
        
    # 4. Sonuçları CPU'ya alıp NumPy formatına çevir
    olasiliklar = olasiliklar.cpu().numpy()
    attention_agirliklari = attention_agirliklari.cpu().numpy()
    
    # 5. Eşik değerine göre kararı ver (0: Yok, 1: Var)
    tahmin_kararlari = (olasiliklar >= esik_degeri).astype(int)
    
    return tahmin_kararlari, olasiliklar, attention_agirliklari

# --- ÖRNEK KULLANIM ---
# Diyelim ki uydudan o anlık tek bir veri paketi geldi (60 zaman adımı, 24 özellik)
ornek_yeni_veri = np.random.randn(60, 24) 

karar, olasilik, dikkat = predict_flare(yuklenen_model, ornek_yeni_veri, esik_degeri=0.5)

print("\n--- ÖRNEK TAHMİN SONUCU ---")
print(f"Modelin Hesapladığı Patlama İhtimali: %{olasilik[0][0] * 100:.2f}")
print(f"Modelin Son Kararı: {'PATLAMA VAR! (1)' if karar[0][0] == 1 else 'Sakin (0)'}")
```

```python
# Hücre 13: Erken Durdurma (Early Stopping) ve Dinamik Öğrenme ile Gelişmiş Eğitim
import torch.optim as optim
import copy # En iyi modeli hafızada tutmak için

# 1. Modeli ve Cihazı Tekrar Hazırla
# Daha önce eğittiğimiz modelin üzerine yazmamak için modeli sıfırlıyoruz
model = SolarFlarePredictor(input_dim=INPUT_DIM, hidden_dim=HIDDEN_DIM, num_layers=NUM_LAYERS).to(device)

# 2. Ceza Katsayısını Biraz Yumuşatıyoruz
yumusatilmis_agirlik = torch.sqrt(gercek_pos_weight).to(device)
print(f"Optimize Edilmiş Yeni Ceza Katsayısı: {yumusatilmis_agirlik.item():.2f}")

# 3. Optimizatör ve Scheduler (Vites Küçültücü)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# HATA ÇÖZÜLDÜ: verbose=True parametresi kaldırıldı
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

# 4. Erken Durdurma (Early Stopping) Ayarları
EPOCHS = 30
EARLY_STOPPING_PATIENCE = 5 
en_iyi_val_loss = float('inf')
iyilesmeyen_epoch_sayisi = 0
en_iyi_model_agirliklari = copy.deepcopy(model.state_dict()) 

print("\n--- AKILLI OPTİMİZASYON İLE EĞİTİM BAŞLIYOR ---")

for epoch in range(EPOCHS):
    model.train()
    toplam_train_loss = 0
    
    # --- EĞİTİM (TRAIN) ---
    for batch_X, batch_y in train_loader:
        batch_X, batch_y = batch_X.to(device), batch_y.to(device)
        optimizer.zero_grad()
        
        tahminler, _ = model(batch_X)
        loss = criterion(tahminler, batch_y)
        
        agirlikli_loss = loss * torch.where(batch_y == 1, yumusatilmis_agirlik, torch.tensor(1.0).to(device))
        ortalama_loss = agirlikli_loss.mean()
        
        ortalama_loss.backward()
        optimizer.step()
        toplam_train_loss += ortalama_loss.item()
        
    model.eval()
    toplam_val_loss = 0
    with torch.no_grad():
        for val_X, val_y in val_loader:
            val_X, val_y = val_X.to(device), val_y.to(device)
            val_tahminler, _ = model(val_X)
            v_loss = criterion(val_tahminler, val_y)
            toplam_val_loss += v_loss.item()
            
    avg_train_loss = toplam_train_loss / len(train_loader)
    avg_val_loss = toplam_val_loss / len(val_loader)
    
    # Güncel öğrenme oranınını al
    guncel_lr = optimizer.param_groups[0]['lr']
    
    print(f"Epoch {epoch+1:02d}/{EPOCHS} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f} | LR: {guncel_lr:.6f}")
    
    scheduler.step(avg_val_loss)
    
    # 6. Erken Durdurma Kontrolü
    if avg_val_loss < en_iyi_val_loss:
        en_iyi_val_loss = avg_val_loss
        iyilesmeyen_epoch_sayisi = 0
        en_iyi_model_agirliklari = copy.deepcopy(model.state_dict())
        print(f"Yeni en iyi model bulundu! (Val Loss: {en_iyi_val_loss:.4f})")
    else:
        iyilesmeyen_epoch_sayisi += 1
        print(f"İyileşme yok. (Kalan Sabır: {EARLY_STOPPING_PATIENCE - iyilesmeyen_epoch_sayisi})")
        
    if iyilesmeyen_epoch_sayisi >= EARLY_STOPPING_PATIENCE:
        print("\nEARLY STOPPING TETİKLENDİ Model ezberlemeye başlamadan eğitim kesildi.")
        break

model.load_state_dict(en_iyi_model_agirliklari)
print("\n Eğitim tamamlandı. En başarılı model ağırlıkları sisteme yüklendi")
```
